```python
import pandas as pd
data = pd.read_csv('stroke.csv')
data.columns = ['id','gender', 'age', 'hypertension', 'heart disease', 'marriage', 'work type','residence type',
                'glucose level','bmi','smoking status','stroke']
data.head(10)
```

```python
data = data.drop(['id'],axis=1)
data = data.drop(['gender'],axis=1)
data = data.drop(['hypertension'],axis=1)
data = data.drop(['heart disease'],axis=1)
data = data.drop(['marriage'],axis=1)
data = data.drop(['work type'],axis=1)
data = data.drop(['residence type'],axis=1)
data = data.drop(['smoking status'],axis=1)
data.head(10)
```

```python
data = data.dropna()
data.head(10)
```

```python
X = data.iloc[:,0:3].values
y = data['stroke']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
TFmodel = confusion_matrix(y_test, y_pred)
print(TFmodel)
print(classification_report(y_test, y_pred))
```
