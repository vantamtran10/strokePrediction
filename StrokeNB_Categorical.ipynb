```python
import pandas as pd
data = pd.read_csv('stroke.csv')
data.columns = ['id','gender', 'age', 'hypertension', 'heart disease', 'marriage', 'work type','residence type',
                'glucose level','bmi','smoking status','stroke']
```

```python
data = data.drop(['id'],axis=1)
data = data.drop(['age'],axis=1)
data = data.drop(['glucose level'],axis=1)
data = data.drop(['bmi'],axis=1)
data = data.drop(['smoking status'],axis=1)
data = data.drop(['work type'],axis=1)
data = data.drop(['residence type'],axis=1)
data.head(10)
```

```python
data['marriage'] = data['marriage'].map({'Yes':1, 'No':0}) 
#data['residence type'] = data['residence type'].map({'Urban':1, 'Rural':0})
#data['work type'] = data['work type'].map({'children':0, 'Govt_jov':1, 'Never_worked':2, 'Private':3, 'Self-employed':4})
#data['smoking status'] = data['smoking status'].map({'never smoked':0, 'formerly smoked':1, 'smokes':2, 'Unknown':3})
data['gender'] = data['gender'].map({'Male':0, 'Female':1, 'Other':2})
data = data.dropna()
data.head(10)
```

```python
X = data.iloc[:,0:4].values
y = data['stroke']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.naive_bayes import CategoricalNB
model = CategoricalNB()
```

```python
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
TFmodel = confusion_matrix(y_test, y_pred)
print(TFmodel)
print(classification_report(y_test, y_pred))
```
